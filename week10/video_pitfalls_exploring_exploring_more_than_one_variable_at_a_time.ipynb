## Choices Explode with Multiple Input Variables


```python
import math

import matplotlib.pyplot as plt

combinations = [math.comb(8, k) for k in range(9)]
plt.bar(range(9), combinations, label="choices")
plt.plot([sum(combinations[i] for i in range(k + 1)) for k in range(9)], color="orange", label="cumulative choices")
plt.legend()
plt.xlabel("Abalone Variables Allowed")
plt.ylabel("Number of Variable Choices")
None
```

## Practical Handling

* 1 variable
  * usually practical unless lots of columns
  * easy to visualize
* 2 variables
  * sometimes practical with a small number of columns
  * easy to visualize
* 3 variables
  * rarely worth systematic coverage
  * hard to visualize
