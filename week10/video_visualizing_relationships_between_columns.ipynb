# Video: Visualizing Relationships Between Columns

This video shows how to visualize correlations and other dependencies between columns.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone.tsv", sep="\t")
abalone
```

```python
_ = pd.plotting.scatter_matrix(abalone)
```

```python
abalone.corr(numeric_only=True)
```

```python
_ = plt.matshow(abalone.corr(numeric_only=True))
```
