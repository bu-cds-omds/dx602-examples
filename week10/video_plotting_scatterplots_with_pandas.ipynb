# Video: Plotting Scatterplots with Pandas

This video shows off pandas support for making scatter plots easily.
Scatter plots are also on the short list with dedicated pandas support, and the `scatter_matrix` function gives a great way to explore connections between all pairs of variables if you have the space to see them.

```python
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone.tsv", sep="\t")
abalone
```

```python
_ = abalone.plot.scatter("Length", "Diameter")
```

```python
_ = abalone.plot.scatter("Length", "Diameter", alpha=0.05)
```

```python
_ = abalone.plot.scatter("Length", "Rings", alpha=0.05)
```

```python
_ = pd.plotting.scatter_matrix(abalone)
```

```python
print(pd.plotting.scatter_matrix.__doc__)
```

```python
_ = pd.plotting.scatter_matrix(abalone, alpha=0.02)
```
