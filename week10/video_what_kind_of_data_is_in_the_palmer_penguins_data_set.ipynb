## Slide: What Columns are in the Data?

| Column Name | Column Type | Column Meaning | Example Values |
|---|---|---|---|
| studyName | string | which research study collected data | PAL0708 <br> PAL0809 <br> PAL0910 |
| Species | string | name of penguin species | Adelie Penguin (Pygoscelis adeliae) <br> Gentoo penguin (Pygoscelis papua) <br> Chinstrap penguin (Pygoscelis antarctica) |
| Region | string | general area of data collection| Anvers |
| Island | string | specific island of data collection | Torgersen <br> Biscoe <br> Dream |
| Stage | string | status of the nest when found | Adult, 1 Egg Stage |
| Individual ID | string | penguin identifier | N1A1 <br> N1A2 <br> N2A1 |
| Clutch Completion | boolean | were two eggs laid in the nest? | Yes <br> No |
| Date Egg | date | date that egg was laid? | 2007-11-11 <br> 2007-11-16 <br> 2007-11-15|
| Culmen Length (mm) | float64 | length of bill | 32.1-59.6 |
| Culmen Depth (mm) | float64 | thickness of bill | 13.1-21.5 |
| Flipper Length (mm) | float64 | length of flippers | 172-231 |
| Body Mass (g) | float64 | body mass | 2700-6300 |
| Sex | string | Biological sex as determined by genetic testing | M <br> F <br> . |
| Delta 15 N (o/oo) | float64 | nitrogen isotope ratio | 7.632200-10.025440 |
| Delta 13 C (o/oo) | float64 | carbon isotope ratio | -27.018540 - -23.787670 |
| Comments | string | free form text, usually addressing missing data| |
