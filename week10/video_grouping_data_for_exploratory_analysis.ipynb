# Video: Grouping Data for Exploratory Analysis

This video shows another style of exploratory analysis based on grouping data and summary tables instead of charting.

```python
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone.tsv", sep="\t")
abalone
```

## Two Kinds of Reports

* What is valuable?
* What changed?



## What is Valuable?

| What | Volume | Average Value |
|---|---|---|
| good stuff | high | high |
| bad stuff | high | low |

More rows are acceptable, but they will be ignored unless the volume is high.
Ideally, all low volume rows are combined into an "other" group.

```python
abalone.groupby("Sex")["Rings"].mean()
```

```python
abalone.groupby("Sex")["Rings"].agg(["size", "mean", "std"])
```
