# Video: Plotting Input Distributions

This video shows how to quickly plot all the input variable distributions with Pandas.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone.tsv", sep="\t")
abalone
```

```python
abalone.describe()
```

```python
abalone.hist(grid=False)
plt.subplots_adjust(wspace=0.4, hspace=0.9)
```

```python
num_cols=3
num_rows=(len(abalone.columns) + num_cols - 1) // num_cols

plt.figure(figsize=(num_cols*4, num_rows*3.5))
for (i, c) in enumerate(abalone.columns):
    plt.subplot(num_rows, num_cols, i + 1)
    abalone[c].hist(grid=False)
    plt.title(c)
```
