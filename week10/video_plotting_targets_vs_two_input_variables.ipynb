# Video: Plotting Targets vs Two Input Variables

This video shows how to plot the relationship between a target variable and two input variables.


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone.tsv", sep="\t")
abalone
```

```python
_ = pd.plotting.scatter_matrix(abalone, c=abalone["Rings"])
```

```python
_ = abalone.plot.scatter("Viscera_weight", "Shell_weight", c="Rings")
```

```python
_ = abalone.plot.scatter("Viscera_weight", "Shell_weight", c="Rings", colormap='viridis')
```

```python
_ = abalone.plot.scatter("Viscera_weight", "Shell_weight", c="Rings", colormap='viridis', s=1)
```
