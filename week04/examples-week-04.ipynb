Accumulator pattern: building a value iteratively with a loop
<li>set up accumulator variable
<li>process a sequence using a loop
<li>update accumulator variable inside loop body

```python
def add_up(numbers):
    '''add up the elements in a list of numbers'''

    # declare an accumulator variable
    total = 0

    # use a loop to iterate over a sequence
    for n in numbers:

        # inside the loop: update the accumulator variable
        total = total + n

    # copy the value of total out of the function
    return total

result = add_up(numbers)
print("result =", result)
```

Computing factorials

```python
def factorial(n):
    '''return the factorial of n
    fact(n) = n * (n-1) * (n-2) * ... * (1)'''
    # declare an accumulator variable
    fact = 1

    # use a loop to iterate over a sequence
    for i in range(n,0,-1):
        # print(f'i = {i}')
        # inside the loop: update the accumulator variable
        fact = fact * i

    return fact

print("factorial(3) =", factorial(3))
print("factorial(7) =", factorial(7))
```

Accumulating a list<br>
Example: reciprocal numbers



```python
def reciprocal_numbers(n):
    '''
    return the list of numbers [1, 1/(1+1), 1/(1+2), 1/1+3), ...]
    '''
    # declare an accumulator variable
    result = []
    
    # use a loop to iterate over a sequence
    for i in range(n):
        # print(f'i = {i}')
        # inside the loop: update the accumulator variable
        r = 1 / (1 + i)
        # print(f'i = {i}, r = {r}')
        result.append(r)
        # result = result + [r] 

    return result


print('reciprocal_numbers(5) =', reciprocal_numbers(5))
```

Accumulating a list<br>
Example: Fibonacci numbers



```python
def fibonacci(n):
    '''Return the first n Fibonacci numbers.
    [0, 1, 1, 2, 3, 5, 8, ...]
    '''
    # declare an accumulator variable
    result = [0, 1]
    
    # use a loop to iterate over a sequence
    for i in range(2, n):

        # inside the loop: update the accumulator variable
        f = result[-2] + result[-1] 
        result.append(f)

    return result

# test case:
print('fibonacci(20) =', fibonacci(20))
```

ASCII Codes, Converstion

```python
## helper functions to convert from character to ascii number and back
# print(ord('q'))
# print(chr(113))
def string_to_ascii(s):
    '''convert a string `s` into a list of ASCII codes.'''

    # set up an accumulator variable:
    result = []

    # use a loop to interate over a sequence:
    for c in s:

        # convert to ASCII code
        n = ord(c)
        # update the accumulator
        result.append(n)

    return result

def ascii_to_string(codes):
    '''convert a list of ASCII codes into a string, 
    and return the string'''

    # set up an accumulator variable:
    result = '' # start with an empty string

    # use a loop to interate over a sequence:
    for x in codes:
        # convert to character:
        c = chr(x)
        # update the accumulator
        result = result + c # appending one more character to the string

    return result

# print('string_to_ascii("hello")', string_to_ascii("hello"))
print("ascii_to_string([104, 101, 108, 108, 111])", ascii_to_string([104, 101, 108, 108, 111]))
```
