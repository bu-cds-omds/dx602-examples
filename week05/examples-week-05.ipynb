Week 5 Examples: Dictionaries (a.k.a. "maps")
<li>associatative collection of key-value pairs
<li>keys must be "hashable"
<li>order is not maintained
<li>values can be any data type, including other objects (e.g., int, string, list, dict)

```python
# drinks = {} # empty dictionary (dict)
# drinks['coffee black'] = 12 # inserting a key/value pair
# drinks['tea'] = 17

# initializing a non-empty dict
drinks = {'coffee black': 12, 'tea': 17}
drinks['diet-coke'] = 4
drinks['seltzer'] = 9
print(drinks)

# extract a value using the key:
# print("drinks['coffee black']", drinks['coffee black'])

# k = 'diet-coke'
# print('drinks[k]', drinks[k])

# use a loop to iterate over a dict
# for k in drinks:
#     print(k,'->', drinks[k])

# obtain all of the keys
keys = list(drinks.keys())
keys.sort() # sort based on natural ordering
# print(keys)
for k in keys:
    print(k,'->', drinks[k])

# values = drinks.values()
values = [v for v in drinks.values()]
print(values)
```

Using dictionary as an accumulator

```python
# process a list, count occurences
import random
# items = [random.randint(1,10) for i in range(10)]
items = [c for c in 'the long and winding road']
print('items =',items)

# count how many times each value occurs:
counts = {}

# go through each item, and use that item as key
for x in items:

    # if we've seen it before, take the old count + 1
    if x in counts:
        counts[x] = counts[x] + 1
    # if this is the first time we see this item, set it's count to 1
    else:
        counts[x] = 1

# print out:
# print('counts =',counts)
keys = list(counts.keys())
keys.sort() # sort based on natural ordering
for k in keys:
    print(k,'->', counts[k], counts[k] * '*')

```

Additional dictionary practice

```python
# combine two dictionaries
dict1 = {'a': 5, 'b':1}
dict2 = {'b': 4, 'c':6}

# add these dictionaries together
# dict1 + dict2 ## unsupported operand type(s) for +: 'dict' and 'dict'

# find all keys from both dictionaries
combined_keys = list(dict1.keys()) + list(dict2.keys())
combined_keys = set(combined_keys)

# combine both dictionaries into one result
result = { k: 0 for k in combined_keys}

# iterate through all keys:
for k in combined_keys:
    # grab values from each dictionary, put into result dict
    if k in dict1:
        result[k] = result[k] + dict1[k] # result[k] += dict1[k]
    if k in dict2:
        result[k] = result[k] + dict2[k]

print('dict1', dict1)
print('dict2', dict2)

# copy = result # didn't create a copy; it copied the reference to the variable result
copy = { k: result[k] for k in result} # create "deep" copy of result
copy['d'] = 42 # add an element
print('result', result)
print("copy", copy)
```
