# Video: Keep Your Colors and Symbols Consistent

This video illustrates the confusion of mixing up colors across charts, and implores you to take measures to keep them consistent.

* Let us take a look at some charts for one of my favorite subjects, which fruit tastes best.

```python
import random
import matplotlib.pyplot as plt
```

```python
days = [1, 2, 3, 4, 5, 8, 9, 10, 11, 12]
mango_scores = [random.random() * 2 + 3 for _ in days]
pear_scores = [random.random() * 2 + 2 for _ in days]
banana_scores = [random.random() * 2 + 1 for _ in days]
```

* You can probably tell from the formulas for this synthetic data that I like mangos the best.

```python
plt.plot(days, mango_scores, label="mango")
plt.plot(days, pear_scores, label="pear")
plt.plot(days, banana_scores, label="banana")
plt.legend()
```

* So blue mangos on top, then orange pears, and finally green bananas.

* If we look at the lesser fruits, things get weird.

```python
plt.plot(days, pear_scores, label="pear")
plt.plot(days, banana_scores, label="banana")
plt.legend()
```

* You can set the colors manually.

```python
plt.plot(days, pear_scores, label="pear", color="orange")
plt.plot(days, banana_scores, label="banana", color="green")
plt.legend()
```

* But is often difficult to match manually chosen colors with the automatically chosen colors.
* Usually it is cleaner, though a bit tedious, to just set colors every time from the beginning.

```python
plt.plot(days, mango_scores, label="mango", color="blue")
plt.plot(days, pear_scores, label="pear", color="orange")
plt.plot(days, banana_scores, label="banana", color="green")
plt.legend()
```

* This issue just goes away if you are able to just show the same set in every chart.
* If this makes sense in your application, usually that is a better choice.

* By the way, try not to rely too much on colors to distinguish lots of different groups of data.
* It will make things difficult for people who are colorblind.
* Even for people with good color vision, nice color palettes usually run out after 5-10 colors, and everything looks garish and is still hard to read.
