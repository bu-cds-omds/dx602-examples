# Video: Plotting Lines and Points

This video covers the most basic usage of the plot command for line graphs and scatter plots.


```python
import matplotlib.pyplot as plt
```

* The most basic usage of the plot function is passing in sequences of x and y values.

```python
plt.plot([2, 4, 6], [5, 3, 7])
```

* If you run multiple plot commands within a Jupyter cell, or before calling show outside of Jupyter, you will get multiple plots in the same chart.

```python
plt.plot([2, 4, 6], [5, 3, 7])
plt.plot([1, 3, 5, 7], [0, 10, 3, 5])
```

* If you have multiple plots, you should label them.

```python
plt.plot([2, 4, 6], [5, 3, 7], label="first data")
plt.plot([1, 3, 5, 7], [0, 10, 3, 5], label="second data")
plt.legend()
```

* The plot function defaults to plotting lines.
* There are a lot of formatting options that you can pass in after the x and y values.

```python
plt.plot([2, 4, 6], [5, 3, 7], "k*", label="first data")
plt.plot([1, 3, 5, 7], [0, 10, 3, 5], "r-", label="second data")
plt.legend()
```

* The first version of formatting controls reproduces MATLAB's plot format options.
* If you have used MATLAB before, and so has the rest of your team, use them.
* Otherwise, use the keyword arguments to control the formatting like this.

```python
plt.plot([2, 4, 6], [5, 3, 7], color="black", linestyle="none", marker="*", label="first data")
plt.plot([1, 3, 5, 7], [0, 10, 3, 5], color="red", linestyle="solid", label="second data")
plt.legend()
```

* If you turn off the lines, then you can use the plot function for scatter plots.

```python
plt.plot([1, 5, 3, 4, 2], [4, 2, 3, 5, 5], linestyle="none", marker=".")
```

* By the way, the plot function does not care what order that you give it data.
* But generally you'll want to make sure the x values are sorted for most curves that you plot, because it will connect the lines in the order that you give the data.
* Repeating the data from the scatter plot,

```python
plt.plot([1, 5, 3, 4, 2], [4, 2, 3, 5, 5], marker=".")
```

* While if you sort the x values, it looks like this.

```python
plt.plot([1, 2, 3, 4, 5], [4, 5, 3, 5, 2], marker=".")
```

* That is it for our introduction to the plot function.

**Code Notes:**
* The [Matplotlib function `matplotlib.pyplot.plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html), referenced as `plt.plot`, takes in a varying number of arguments based on how much control you need of the chart.
  * The first arguments are for the $x$ and $y$ values of the data to be plotted.
    * Most commonly, $x$ and $y$ are provided as sequences of numbers of the same length. (Next week, we will call these 1-dimensional array-like objects.)
    * $x$ values can be omitted, just passing $y$ values, in which case the $x$ values will be assumed to be $0, 1, 2, \ldots$ However we strongly encourage passing the $x$ values explicitly. Even if you just pass a range object, that will make it clear what you intended.
    * Another less common option is to pass a 2-dimensional array with $x$ and $y$ columns. We won't use that option much either, but it is occasionally convenient.
  * After the $x$ and $y$ values, the next argument is to control the formatting of this line.
    * This is an idiosyncratic convention from MATLAB.
    * The format argument is a one to three character string that can indicate color, line style, and marker following MATLAB's usage.
    * We only recommend using this format argument if you and your team are already familiar with MATLAB plot formatting. Otherwise, use the keyword arguments below for clarity.
  * Use the optional `color` argument to set the color of each line.
  * Use the optional `linestyle` argument to control how each line is drawn. The video had examples using "none" and "solid". "none" will remove the line, just leaving the markers. See [this documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle) for the full list of `linestyle` choices.
  * Use the optional `marker` argument to control the markers drawn at each data point. The default is no marker. See [the marker documentation](https://matplotlib.org/stable/api/markers_api.html) for the full list of choices.
  * Use the optional `label` argument whenever plotting more than one set of data in the same chart.
* The [Matplotlib function `matplotlib.pyplot.legend`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) will add a legend to the current chart in progress.
  * Otherwise, labels will not be shown even if you include them in your `plot` calls.
  * An optional `loc` argument lets you control where the legend is added. Generally, it will make a decent choice on its own and you will not have to set this explicitly.
* The [Matplotlib function `matplotlib.pyplot.scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) can also generate scatter plots, and with very granular control over individual points being plotted. `scatter` can control the marker, color, and size of each point plotted. Most of the time, you won't need this level of control, so using `plot` will be more efficient. An example of when you might use `scatter` for data science would be varying the marker size to reflect sample weights (number of samples).
