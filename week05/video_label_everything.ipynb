The last video covered the `plot` function and its major options.
We will now cover some of the surrounding functions to make sure everything in your chart is clearly labeled.

* We are going to look at Matplotlib functionality to clearly label your charts.
* Let's start with a plain example.

```python
import math
import matplotlib.pyplot as plt

xs = [i / 10 for i in range(100)]
y1s = [math.cos(x) for x in xs]
y2s = [math.sin(x) for x in xs]

plt.plot(xs, y1s)
plt.plot(xs, y2s)
None
```

* We already saw the label argument and legend function in the previous function, so I will add those now.

```python
plt.plot(xs, y1s, label="blood sugar")
plt.plot(xs, y2s, label="excitement")
plt.legend()
None
```

* As you can see, the legend function will look for a spot with more empty space.
* It will have more trouble if the labels are long and there is not a lot of white space in the chart.


```python
plt.plot(xs, y1s, label="blood sugar")
plt.plot(xs, y2s, label="excitement")
plt.legend()
None
```

* Let's add more information to this chart.

```python
plt.plot(xs, y1s, label="blood sugar")
plt.plot(xs, y2s, label="excitement")
plt.legend()
plt.title("Professor Response to Too Much Sugar in Coffee")
plt.xlabel("Time since First Sugar Cube")
plt.ylabel("Measurement Delta from Baseline")
None
```

**Code Notes:**
* The [Matplotlib function `matplotlib.pyplot.legend`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html), referenced as `plt.legend`, will add the legend to the current chart. This legend depends on you using the optional `label` argument when calling `plot`.
* The [Matplotlib function `matplotlib.pyplot.title`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html) will set the title on a chart in progress.
* The [Matplotlib function `matplotlib.pyplot.xlabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html), referenced as `plt.xlabel`, will set the label for the x axis. Similarly, the [Matplotlib function `matplotlib.pyplot.ylabel`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html) will set the label for the y axis.
* The [Matplotlib function `matplotlib.pyplot.annotate`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html), referenced as `plt.annotate`, will let you add text at arbitrary points on the charge. With a few extra arguments, you can add an arrow to clarify which data your text refers to. This can be helpful to address anticipated questions about your charts, so use it thoughtfully.
