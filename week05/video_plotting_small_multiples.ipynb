# Video: Plotting Small Multiples

This video shows an example of plotting data with small multiples, variations of the same chart repeated, usually across different subsets of the data.


```python
!wget https://raw.githubusercontent.com/bu-cds-omds/bu-cds-omds-data/main/data/abalone.tsv
```

```python
import csv
with open("abalone.tsv") as abalone_file:
    abalone_reader = csv.DictReader(abalone_file, dialect="excel-tab")
    abalone_data = list(abalone_reader)

abalone_data[0]
```

```python
sexes = sorted(set(r['Sex'] for r in abalone_data))
sexes
```

```python
for row in abalone_data:
    row["Rings"] = int(row["Rings"])
```

Scripts:
* Let's start plotting now.

```python
import matplotlib.pyplot as plt
```

```python
rows = 3
cols = 1

for (i, sex) in enumerate(sexes):
    plt.subplot(rows, cols, i + 1)
    plt.hist([r["Rings"] for r in abalone_data if r["Sex"] == sex])
    plt.title(f"Sex = {sex!r}")
    plt.xlabel("Rings")
    plt.ylabel("Count")
```

```python
for (i, sex) in enumerate(sexes):
    plt.subplot(rows, cols, i + 1)
    plt.hist([r["Rings"] for r in abalone_data if r["Sex"] == sex])
    plt.title(f"Sex = {sex!r}")
    plt.xlabel("Rings")
    plt.ylabel("Count")

plt.subplots_adjust(hspace=1.0)
```

```python
for (i, sex) in enumerate(sexes):
    plt.subplot(rows, cols, i + 1)
    plt.hist([r["Rings"] for r in abalone_data if r["Sex"] == sex], bins=30, range=(0,30))
    plt.title(f"Sex = {sex!r}")
    plt.xlabel("Rings")
    plt.ylabel("Count")

plt.subplots_adjust(hspace=1.0)
```
