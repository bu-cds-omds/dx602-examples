## Not Starting from Zero Exaggerates Trends

![A financial chart exaggerating improvements to margins by not starting the y axis from zero.](https://github.com/bu-omds/dx602-examples/assets/870970/e7d1d51d-3380-4bfb-87f0-718fe2bc44d8)


```python
# @title ##### Slide: Fixing the Y Axis to Start from Zero

import matplotlib.pyplot as plt

plt.plot([2018, 2019, 2020, 2021, 2022, 2023], [0.282, 0.295, 0.313, 0.310, 0.317, 0.348])
axis = plt.gca()
axis.set_ylim(0, 0.36)
plt.xlabel("Year")
plt.ylabel("Variable Marketing Margin %")
None
```

## A Non-Linear Axis?

![image.png](https://github.com/bu-omds/dx602-examples/assets/870970/e6311886-5b78-403a-838b-2c8b59484220)

```python
# @title ##### Slide: Straightening the Y Axis

import matplotlib.pyplot as plt

#plt.plot([2018, 2023], [0, 0])
plt.plot([2018, 2019, 2020, 2021, 2022, 2023], [-0.033, 0.034, 0.053, 0.035, 0.015, 0.001], label="Adjusted EBITDA Margin %")
plt.grid(True)
plt.gca().get_xaxis().set_visible(True)
plt.legend()
None
```

**Code Notes:**
* The [Matplotlib function `matplotlib.pyplot.gca`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gca.html), referenced as `plt.gca`, returns an axis object.
  * In this code, the axis object's `set_ylim` method is used to set the range of y values.
