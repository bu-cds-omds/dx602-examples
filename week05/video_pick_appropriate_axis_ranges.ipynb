## Not Starting from Zero Exaggerates Trends

![A financial chart exaggerating improvements to margins by not starting the y axis from zero.](https://github.com/bu-cds-omds/dx602-examples/assets/870970/e7d1d51d-3380-4bfb-87f0-718fe2bc44d8)


Script:
* Here is the original chart with all the labels.
* The 2018 percentage is not close to zero at all.
* So this chart has a funky y axis for the margin percentage.
* Why was this done?
* If you do not look at the percentage labels, it looks like the margin percentage started close to zero in 2018, and has increased substantially over the following five years.
* Like the margin was nothing in 2018, and now it is pretty great.
* Good job team?

## Fixing the Y Axis to Start from Zero


```python
import matplotlib.pyplot as plt

plt.plot([2018, 2019, 2020, 2021, 2022, 2023], [0.282, 0.295, 0.313, 0.310, 0.317, 0.348])
plt.gca().set_ylim(0, 0.36)
plt.xlabel("Year")
plt.ylabel("Variable Marketing Margin %")
None
```

Script:
* Here is a chart that I made with just the margin percentage numbers.
* The margin percentage did indeed improve over time.
* We already knew that -- we saw the numbers.
* But the meteoric rise from nothing narrative is gone.
* This chart shows a slow but steady improvement.
* Extending the y axis to include zero keeps those changes in perspective.
* Sometimes you do want to focus in on those smaller changes, but it should be done explicitly and intentionally.

## Straightening the Y Axis


```python
import matplotlib.pyplot as plt

plt.plot([2018, 2019, 2020, 2021, 2022, 2023],
         [-0.033, 0.034, 0.053, 0.035, 0.015, 0.001],
         marker="x")

# three ways to add the line at zero
plt.gca().axhline(y=0, color="k")
# plt.plot([2018, 2023], [0, 0]) # add a line, but uses up a chart color
# plt.grid(True) # show grid

plt.xlabel("Adjusted EBITDA Margin %")
plt.ylabel("Year")

None
```

Script:
* I recreated the margin percentage chart from their numbers again.
* I added an extra line at zero to make the signs and magnitudes clear.
* Finding instructions for that are surprisingly annoying online, and most of the answers on sites like StackOverflow do not work.
* If you look at the code in the notebook linked after this video, you can see the official way to do it, along with two other hacks that I used in the past before finding the official way.
* Let's look at those numbers again.
* Now it is clear that 2018 was significantly negative, but that was reversed in 2019.
* 2020 was better, but then margins dropped to basically zero in 2023.
* The drop from 2020 to 2022 was almost a straight line, so I added markers to each data point to see where 2021 was.
* This is a very different story from looking at the lines in the original chart, and assuming that the bar chart zero was also the zero for the margin percentage lines.
* Sometimes you get charts like these by just plotting without any fine tuning.
* Which makes it important for you take the time to add on the labels and tweak the axis ranges to communicate the truth clearly.
* Those charts had the truth, but it could have been more clear.


**Code Notes:**
* The [Matplotlib function `matplotlib.pyplot.gca`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gca.html), referenced as `plt.gca`, returns an axis object.
  * In this code, the axis object's `set_ylim` method is used to set the range of y values.
