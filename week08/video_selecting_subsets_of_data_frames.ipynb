# Video: Selecting Subsets of Data Frames

This video shows how to flexibly select different sets of rows and columns from a pandas data frame.


## Slide: Selecting Subsets of Data Frames

* Pandas data frames have two dimensions, rows and columns.
* You can select a subset of either them.
* You can reorder them simultaneously.

## Code Example: Selecting Columns

```python
import pandas as pd
```

```python
df = pd.DataFrame({"c1": [0, 3, 6], "c2": [1, 4, 7], "c3": [2, 5, 8]}, index=["r1", "r2", "r3"])
df
```

```python
df[["c1", "c3"]]
```

```python
df[["c2", "c2"]]
```

```python
df[["c3", "c2", "c1"]]
```

```python
df.iloc[[2, 0]]
```

```python
df.iloc[[2, 0], [1, 2]]
```

```python
df.loc[["r1", "r3"]]
```

```python
df.loc[["r1", "r3"], ["c1", "c3"]]
```

```python
df.iloc[0::2]
```
