Numpy arrays

```python
# use a python list to create numpy array
a = np.array([1,2,3,4,5])
b = np.array(range(10, 5, -1))

print(f'a = {a}; b = {b}')

# element-wise arithmetic operations
# c = a + b
# print(f'c = {c}')
# c = a - b
# print(f'c = {c}')
# c = a * b
# print(f'c = {c}')
# c = a / b
# print(f'c = {c}')

# dot-product
c = a @ b
print(f'c = {c}')

# aggregate operations
print(f'sum of a: {a.sum()}')
print(f'min of a: {a.min()}')
print(f'max of a: {a.max()}')
print(f'mean of a: {a.mean()}')
print(f'std of a: {a.std()}')
```

The shape of an array

```python
a = np.array([10,20,30,40])
print('a.shape =', a.shape)
# b = a.reshape((4,1))
b = a.reshape((len(a),1))
print('b.shape =', b.shape)
print(f'a = \n{a}')
print(f'b = \n{b}')
c = a.reshape((2,2))
print(f'c = \n{c}')
```

About array products:

```python
a = np.array(range(0,24)).reshape((4,6))
b = np.array(range(0,24)).reshape((6,4))
print(f'a = \n{a}')
print(f'b = \n{b}')
# c = a * b #element-wise product
c = a @ b #dot-wise product # note: inner dimension must be the same
print(f'c = \n{c}')
```

Transpose of an array

```python
a = np.array(range(0,10)).reshape((2,5))
b = a.transpose() 
print(f'a =\n {a}')
print(f'b =\n {b}')
```
