Create nicely-formatted time, show table

```python
def format_hh_mm_ss(hours):
    '''Convert a value in hours to HH:MM:SS formatting.'''
    h = int(hours)
    m = (hours - h) * 60
    s = (m - int(m)) * 60
    m = int(m)
    s = int(s)
    
    return f"{h:02}:{m:02}:{s:02}"


```
