```python
# general import statements
import matplotlib.pyplot as plt
import numpy as np
```

Basic plots...

```python
numbers = [3,1,4,5,9]
# plt.plot(numbers) # most basic
# define some x values:
x = [1, 2, 3, 4, 5]
plt.plot(x, numbers)
plt.ylabel("value")
plt.xlabel("numbers")
plt.title("we like pie!")

```

Plotting multiple things together

```python
plt.plot(x, numbers, color="blue", marker="*")
plt.bar(x=x, height=numbers, color="#9400D3")
```

Some waves

```python
# generate some x values:
x = np.arange(0,10,0.1)
# print(x)
y = np.sin(x)
plt.plot(x, y)
plt.title("catch a wave and you're sitting on top of the world")
plt.ylabel("sin(x)")
plt.xlabel("x-values")
```

Two waves...

```python
# generate some x values:
x = np.arange(0,10,0.1)
# print(x)
y1 = np.sin(x)
y2 = np.cos(x)

# call plot 2 times:
# plt.plot(x, y1)
# plt.plot(x, y2)

# a 2-d array with sin(x) and cos(x)
ys = np.array([y1, y2]) # an array to hold multiple columns
print(f'x.shape={x.shape}')
print(f'ys.T.shape={ys.T.shape}')
plt.plot(x, ys.T) # plot the transpose of ys

plt.title("catch a wave and you're sitting on top of the world")
# plt.ylabel("sin(x)")
plt.xlabel("x-values")
plt.legend(['sin(x)', 'cos(x)'])
```

Scatter plot

```python
# generate some x values:
x = np.arange(0,10,0.1)
# print(x)
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x, y1)
plt.title('scatter plot of x, sin(x)')
```

Bar chart

```python
# generate some x values:
x = np.arange(0,10,0.1)
# print(x)
y1 = np.sin(x)
y2 = np.cos(x)

plt.bar(x, y1)
plt.title('bar chart of x, sin(x)')
```

Random numbers.... (uniform distribution)

```python
# drawing random numbers from uniform distribution
random_numbers = np.random.random(10000)

# make a histogram of these:
_ = plt.hist(random_numbers, bins=20) # returns (counts, column_boundaries, plot object)
```

Random numbers (normal distribution)

```python
# draw some random numbers from the normal distribution:
n = 10000
z = np.random.normal(size=n)
_ = plt.hist(z, bins=100)
plt.xlabel("number of standard deviations")
plt.ylabel("count")
```

```python
# draw some random numbers from the normal distribution:
n = 100000
z = np.random.normal(size=n)
counts, bin_edges, _ = plt.hist(z, bins=100)
plt.xlabel("number of standard deviations")
plt.ylabel("count")

from scipy.stats import norm
x_axis = np.arange(-3, 3, 0.01)
y_axis = norm.pdf(x_axis, 0, 1) # mean of 0, std of 1
y_max = y_axis.max() # numpy max value


# calculate the width of each bin:
bin_width = bin_edges[1] - bin_edges[0]
print(f'bin_width={bin_width}')
# scale up the std normal curve
y_axis *= (n * bin_width)
plt.plot(x_axis, y_axis, color="red")
plt.title("this is pretty normal")

```

Plotting multiple graphs together

```python
# fig, ax = plt.subplots(1, 2) # rows, columns
fig, ax = plt.subplots(1, 2, figsize=(10, 5)) # rows, columns, figsize=(width, height)
# plot in each of these axes:
ax[0].plot(x_axis, y_axis, color="red") # left hand side
ax[0].set_title("this is normal")
ax[1].hist(z, bins=50) # right hand side
ax[1].set_title("this is approximately normal")
```

Pie chart

```python
labels = ["coffee black", "tea", "macchiato", "mocha"]
counts = [17, 4, 6, 11]
# _ = plt.pie(counts, labels=labels)

fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].pie(counts, labels=labels)
ax[0].set_title("what we drink")
ax[1].bar(x=labels, height=counts)
ax[1].set_title("we love pie")
```
