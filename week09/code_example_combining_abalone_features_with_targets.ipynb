# Code Example: Combining Abalone Features with Targets

The abalone data set that we've used a few times is actually the combination of two datasets about the same abalone.
The first dataset is physical measurements taken of the abalone.
The second dataset is counts of the rings for the same abalone.
The latter dataset is collected by grinding the shells carefully and examining the cross sections under the microscope, so this data is difficult to collect.
The plan of the research project was to use the physical measurements to predict the rings, and then predict the abalone ages using the estimate $Rings + 1.5$ that had previously been derived.

```python
import pandas as pd
```

```python
abalone_features = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone-features.tsv", sep="\t")
abalone_features
```

```python
abalone_targets = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/main/data/abalone-targets.tsv", sep="\t")
abalone_targets
```

These files had the same abalone in the same order, so the default range index works to match up data.
After loading each file, they can be combined easily by just copying the columns.
Here are three examples, copying the data one column at a time, copying all the columns at once, and using the join function.

These examples will start by copying `abalone_features` to keep it untouched for the following examples.
Whether this is necessary will depend on your use case.

```python
abalone_combined_1 = abalone_features.copy()
abalone_combined_1["Rings"] = abalone_targets["Rings"]
abalone_combined_1
```

The next version copies all the columns at once.

```python
abalone_combined_all = abalone_features.copy()
abalone_combined_all[abalone_targets.columns] = abalone_targets
abalone_combined_all
```

The last version using joins behaves differently if the index values are not the same.
Using `join()` with the default options, the result will be rows whose index value appeared in both data frames.
The previous versions that added columns had the same rows as the original data frame, but NA values where the added columns were missing the index value.

```python
abalone_combined_join = abalone_features.join(abalone_targets)
abalone_combined_join
```

All three cases were pretty easy and expressable with one-liners if you are not concerned about preserving the original data frame.
