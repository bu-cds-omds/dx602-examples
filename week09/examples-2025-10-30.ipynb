```python
# these lines were copied from homework 9
import pandas as pd
penguins_adelie = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/refs/heads/main/data/palmer-penguins-adelie.csv", index_col="Sample Number")
penguins_gentoo = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/refs/heads/main/data/palmer-penguins-gentoo.csv", index_col="Sample Number")
penguins_chinstrap = pd.read_csv("https://raw.githubusercontent.com/bu-cds-omds/dx602-examples/refs/heads/main/data/palmer-penguins-chinstrap.csv", index_col="Sample Number")
penguins = pd.concat([penguins_adelie, penguins_gentoo, penguins_chinstrap])
penguins.head()
```

```python
# print(penguins.columns)
# print(penguins['Body Mass (g)'].describe())
# penguins['Body Mass (g)'].hist(bins=20, legend=True)
# print(penguins['Species'].describe())
# print(penguins['Species'].value_counts())
# print(penguins['Island'].value_counts())

# group by one column (categorical data)
# grouping = penguins.groupby('Island')
# print(grouping['Body Mass (g)'].describe())
# grouping['Body Mass (g)'].hist(legend=True)

grouping = penguins.groupby(['Island', 'Species'])
# print(grouping['Body Mass (g)'].describe())
# grouping['Body Mass (g)'].hist(legend=True)
df_sex_by_group = grouping['Sex'].value_counts()
# print(df_sex_by_group)

# pivot table example:
# index is the index for the output of this table
# columns is the set of columns to generate in the output
# values is used for the aggregate function
# penguins.pivot_table(index='Species', columns='Sex', values='Body Mass (g)')
print("min body mass by category:")
print(penguins.pivot_table(index='Island', columns='Sex', values='Body Mass (g)', aggfunc='min'))
print("max body mass by category:")
print(penguins.pivot_table(index='Island', columns='Sex', values='Body Mass (g)', aggfunc='max'))
print("count body mass by category:")
print(penguins.pivot_table(index='Island', columns='Sex', values='Body Mass (g)', aggfunc='count').fillna(0))
```

```python
# these data files from assignment 9
df_components = pd.read_csv('components_15.tsv', sep='\t')
df_orders = pd.read_csv('orders_15.tsv', sep='\t')
# print(df_components)
# print(df_orders)

# df.join works when both dataframes share the same index
# df_component_orders = df_components.join(df_orders, on="device")
# df.merge works when not same index, but they do have a common column
df_component_orders = df_components.merge(df_orders, on="device")

# how many of each component do we need to order?
df_component_orders['order_qty'] = df_component_orders['component_quantity'] * df_component_orders['device_quantity']
print(df_component_orders)

# what is the total quantity of stuffs we need to be able to fulfill these orders?
print('*' * 50)
grouping_by_component = df_component_orders.groupby('component')
print(grouping_by_component['order_qty'].sum())

```
