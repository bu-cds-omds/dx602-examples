Reading a TSV (tab separated values) structured file

```python
# sample TSV file: 
filename = 'data6_a.tsv'
with open(filename) as f:

    # lines = f.readlines()
    # print(lines) # note the embedded '\t' character for tabs

    # read headers into a list:
    # headers = f.readline().strip().split() # default: split on white space
    headers = f.readline().strip().split('\t') # split on tab character
    print(headers)

    # process the rest of the file:
    for line in f:

        # extract a list of fields:
        values = line.strip().split('\t')
        # print(values)
        
        # build a list of lists:
        # result = [ [ headers[i], values[i]] for i in range(len(headers)) ]
        # print(result)
        # build a dictionary
        d = { headers[i] : values[i] for i in range(len(headers)) }
        print(d)

```

Using a dictionary for counting words

```python
filename = "time.txt"

# read the entire file as a string, convert to lower case
# and split to a list
with open(filename) as f:

    text = f.read()
    words = text.strip().lower().split() # default delimiter is white space
    # print(words)

    counts = {} # empty dictionary
    # iterate of all words in the list, and count them
    for w in words:
        # if we've seen this word w before, add to the count
        if w in counts:
            c = counts[w]
            counts[w] = c + 1
        
        # this is the first time we see this word
        else:
            counts[w] = 1
    # done processing all words
    # print(counts)
    # obtain a list of k-v pairs:
    pairs = list(counts.items())
    pairs.sort(key=lambda x: x[1], reverse=True)
    # print(pairs)
    # print out the top 10 words and counts:
    for pair in pairs[:10]:
        print(f'{pair[0]}\t\t{pair[1]}')

    # save result as JSON file:
    output_filename = "word_counts.json"
    with open(output_filename, "w") as f_out:
        
        # store the data (dictionary) to a JSON file
        json.dump(counts, f_out)

# print out the counts
```
