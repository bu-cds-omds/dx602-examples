Write a function named `traffic_fine(speed, limit)` that:

<li> Accepts two integers: speed (vehicle speed in km/h) and limit (speed limit in km/h).
<li> Returns the appropriate fine string based on how much the speed exceeds the limit:
<li> If speed is less than or equal to limit: return "No fine".
<li> If speed is 1–20 km/h over the limit: return "Fine: $100".
<li> If speed is 21–40 km/h over the limit: return "Fine: $250".
<li> If speed is more than 40 km/h over the limit: return "Fine: $500 and license suspension".
<li> If either input is negative, return "Invalid input".

```python
def traffic_fine(speed, limit):
    '''
    Accepts two integers: speed (vehicle speed in km/h) and limit (speed limit in km/h).
    Returns the appropriate fine string based on how much the speed exceeds the limit:
    If speed is less than or equal to limit: return "No fine".
    If speed is 1–20 km/h over the limit: return "Fine: $100".
    If speed is 21–40 km/h over the limit: return "Fine: $250".
    If speed is more than 40 km/h over the limit: return "Fine: $500 and license suspension".
    If either input is negative, return "Invalid input".
    '''
    if speed < 0 or limit < 0:
        return "Invalid input"
    elif speed <= limit:
        return "No fine"
    
    # calculate overage first
    over = speed - limit
    if over < 20:
        return "Fine: $100"
    elif over < 40:
        return "Fine: $250"
    elif over >= 40:
        return "Fine: $500 and license suspension"


# test cases:
print(f"traffic_fine(45, 50), {traffic_fine(45, 50)}")
print(f"traffic_fine(55, 50), {traffic_fine(55, 50)}")
print(f"traffic_fine(75, 50), {traffic_fine(75, 50)}")
print(f"traffic_fine(95, 50), {traffic_fine(95, 50)}")
print(f"traffic_fine(-5, 50), {traffic_fine(-5, 50)}")
```


**List comprehension**
<li>Compact syntax to create a list from any sequence (e.g., string, list, range, dict)
<li>Embedded for loop to iterate over sequence
<li>Optional if/else logic



```python
# examples: processing a sequence: strings
letters = 'abc'
# lc = [c*2 for c in letters]
print('[c*2 for c in letters] =', [c*2 for c in letters])
print('[ord(c) for c in letters] =', [ord(c) for c in letters]) #ASCII codes

# list examples
numbers = [6,1,7,3,5,8,2,3,6,1]
print('[x ** 2 for x in numbers] =', [x ** 2 for x in numbers])
evens = [x for x in numbers if x % 2 == 0 ]
print('evens = ', evens)

# processing a range ## range(start, stop, step)
# numbers = [i for i in range(10)]
# numbers = [i for i in range(5, 15)]
# numbers = [i for i in range(5, 15, 2)]
numbers = [i for i in range(15, 5, -2)]
print('numbers =', numbers)

squares = [i * i for i in range(-10, 11)]
print(f'squares = {squares}')
# we will come back to matplotlib in week 9
import matplotlib.pyplot as plt
plt.plot(squares)
```

Optimization using a list of lists
<li> start with a list of options
<li> need a function to "score" each option
<li> build list of sublists
<li> pick the best one using max()

```python
# helper function:
def count_vowels(word):
    '''process a string and return the number of vowels'''

    ones = [1 for c in word if c in 'aeiou']
    return sum(ones)

words = ['welcome', 'to', 'the', 'jungle']
# which has the most vowels?
# print("count_vowels('welcome') returned", count_vowels('welcome'))

# scored_words = [count_vowels(w) for w in words]
scored_words = [[count_vowels(w), w] for w in words]
print("scored_words =", scored_words)

# pick the min or max
best = max(scored_words) # returns a sublist [score, word]
print('best =', best)

# extract the element we're looking for
word = best[-1] # use indexing to extract the element last position
print('word =', word)

```

Another example: which number from 1 to 100 has the most factors?

```python
def factors(n):
    '''
    discover the factors of n -- 
    we can divide n by i with no remainder
    '''
    return [i for i in range(1, n//2+1) if n % i == 0 ]

def num_factors(n):
    '''
    count up the number of factors of n
    '''
    return sum([1 for d in factors(n)]) + 1

def most_factors(numbers):
    '''
    use the list of lists technique find
    the number in the list numbers with the most 
    factors (could be a tie)
    '''
    # score the options
    scored = [[num_factors(n),n] for n in numbers] 

    # pick the best one with max
    best_pair = max(scored)
    #print(best_pair, factors(best_pair[1]))

    # return the number
    return best_pair[1]

# print('factors(12) returned', factors(12))
# print('num_factors(12) returned', num_factors(12))
# which number from 1 to 100 has the most factors?
print(f"most_factors(range(1,100)) returned {most_factors(range(1,100))}")
# print(f"most_factors(range(1,1000)) returned {most_factors(range(1,1000))}")
# print(f"most_factors(range(1,10000)) returned {most_factors(range(1,10000))}")



```
