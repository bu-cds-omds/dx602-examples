Discuss/demonstrate: the difference between print and return

```python
def function_a(x):
    result = x + 1
    print(result)

def function_b(x):
    result = x + 1
    return result

# call the functions to test:
# function_a(3)
# function_b(3)

# when we print a value from a function, we cannot use it elsewhere in our program
x = function_a(3)
# when we return a value from a function, we can use it elsewhere
y = function_b(3)
print('x =', x)
print('y =', y)

# z = x + 2 ## error: we cannot use None in a follow-on computation
z = y + 2
```

Lists are mutable data types

```python
def add_one(x):
    x = x + 1
    return x

# test:
# x = 41
# # print(add_one(x))
# print("x =", x, "id = ", id(x))
# x = x + 1
# print("x =", x, "id = ", id(x))
# print("id(41) = ", id(41))

items = [1, 2, 3, 4]
# print("id(items) =", id(items))
# # re-assigning an element of the list
# items[0] = 99
# print(items)
# print("id(items) =", id(items))

def add_two(items):
    '''Lists are mutable. 
    We can modify the elements, even from within a function.'''
    items[0] = items[0] + 2
    # items[0] += 2 # compound operator: addition and reassignment

# test this function:
print('items =', items)
add_two(items)
print('items =', items)

```

How to specify multiple number of parameters:

```python
# we defined the same function name twice.
# the second defition replaces (overrides) the first one
# def myst(a):
#     return a + 1

# def myst(a, b):

#     return a + b + 1

# solution: default parameters: the parameter b has a default value of 0
def myst(a, b = 0):

    return a + b + 1

# test:
print('myst(2,3) returned', myst(2,3))
print('myst(2) returned', myst(2))
```
